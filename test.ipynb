```python
d = pandas.read_csv("./data/sleepstudy.csv")
```

```python
s = d[["Subject", "Days"]]
```

```python
s['code'] = LabelEncoder().fit_transform(s.Subject)
```

```python
t = s[["Subject", "code"]].pivot(columns='Subject').reindex()
```

```python
t.values[np.isfinite(t.values)] = 1
t.values[np.isnan(t.values)] = 0
```

```python
v = s[["Subject", "Days"]].pivot(columns='Subject').reindex()
```

```python
v.values[np.isnan(v.values)] = 0
```

```python
pp = pd.concat([t, v], axis=1, sort=True)
```

```python
pp.reindex(sorted(pp.columns, key=lambda x: x[1]), axis=1)
```
